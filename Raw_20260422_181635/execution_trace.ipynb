Generate Figure 3: temporal dynamics for highlighted cell types

```python

suppressPackageStartupMessages({
  library(ggplot2)
  library(dplyr)
  library(tidyr)
})

# Use GSE197341 per-timepoint module scores (full 279-gene set)
# These are the best temporal data we have
ms_197 <- read.csv("/mnt/results/06_cell_death/GSE197341_module_scores_by_timepoint.csv")

# Focus on EC, Neuron, Astrocyte
focus_cts <- c("Endothelial","Neuron","Astrocyte")
ms_focus <- ms_197 %>%
  filter(cell_type %in% focus_cts) %>%
  mutate(
    timepoint = factor(timepoint, levels=c("Sham","D3","D7","D14","D28")),
    day = case_when(
      timepoint == "Sham" ~ 0,
      timepoint == "D3"   ~ 3,
      timepoint == "D7"   ~ 7,
      timepoint == "D14"  ~ 14,
      timepoint == "D28"  ~ 28
    )
  )

# Compute delta vs Sham for each cell type × pathway
sham_vals <- ms_focus %>%
  filter(timepoint == "Sham") %>%
  select(cell_type, sham_apo=mean_apoptosis, sham_necr=mean_necroptosis, sham_ferr=mean_ferroptosis)

ms_delta <- ms_focus %>%
  filter(timepoint != "Sham") %>%
  left_join(sham_vals, by="cell_type") %>%
  mutate(
    delta_apoptosis   = mean_apoptosis   - sham_apo,
    delta_necroptosis = mean_necroptosis - sham_necr,
    delta_ferroptosis = mean_ferroptosis - sham_ferr
  ) %>%
  pivot_longer(cols=starts_with("delta_"),
               names_to="pathway", values_to="delta") %>%
  mutate(
    pathway = sub("delta_", "", pathway),
    pathway = tools::toTitleCase(pathway),
    pathway = factor(pathway, levels=c("Apoptosis","Ferroptosis","Necroptosis")),
    cell_type = factor(cell_type, levels=c("Endothelial","Neuron","Astrocyte"))
  )

# Colors
ct_colors <- c("Endothelial"="#0279EE", "Neuron"="#FF9400", "Astrocyte"="#75A025")
pathway_lty <- c("Apoptosis"="solid", "Ferroptosis"="dashed", "Necroptosis"="dotted")

p <- ggplot(ms_delta, aes(x=day, y=delta, color=cell_type, linetype=pathway, group=interaction(cell_type, pathway))) +
  geom_hline(yintercept=0, color="grey60", linewidth=0.4) +
  geom_line(linewidth=0.9) +
  geom_point(size=2.5) +
  scale_color_manual(values=ct_colors, name="Cell type") +
  scale_linetype_manual(values=pathway_lty, name="Pathway") +
  scale_x_continuous(breaks=c(3,7,14,28), labels=c("D3","D7","D14","D28")) +
  labs(
    title    = "Temporal Dynamics of Cell Death Pathways Post-Stroke",
    subtitle = "GSE197341 (dMCAO mouse model, D3–D28)\nDelta = mean module score vs. Sham control",
    x        = "Days post-stroke",
    y        = "Delta module score (stroke \u2212 Sham)"
  ) +
  facet_wrap(~pathway, ncol=3, scales="free_y") +
  theme_classic(base_size=11) +
  theme(
    plot.title    = element_text(face="bold", size=12),
    plot.subtitle = element_text(size=8, color="grey40"),
    strip.text    = element_text(face="bold", size=10),
    legend.position = "right"
  )

ggsave("/workspace/fig3_temporal_dynamics_v2.png", p, width=12, height=5, dpi=200)
system("cp /workspace/fig3_temporal_dynamics_v2.png /mnt/results/06_cell_death/fig3_temporal_dynamics_v2.png")
cat("Figure 3 saved\n")
```

Build GSE197341 Seurat object with cell type annotations and run per-timepoint AverageExpression

```python

suppressPackageStartupMessages({
  library(Seurat)
  library(Matrix)
  library(dplyr)
})

# mat, meta197 already loaded above
cat("Matrix in memory:", nrow(mat), "x", ncol(mat), "\n")

# ── Create Seurat object ───────────────────────────────────────────────────────
cat("Creating Seurat object...\n")
seu <- CreateSeuratObject(counts=mat, min.cells=3, min.features=200)
cat("After filtering:", ncol(seu), "cells,", nrow(seu), "genes\n")

# Add timepoint metadata
samples_vec <- meta197[colnames(seu), "Samples"]
tp_vec <- sub(" .*", "", samples_vec)   # "D3 1" -> "D3"
seu@meta.data$timepoint <- tp_vec
seu@meta.data$condition <- ifelse(tp_vec == "Sham", "Sham", "Stroke")
cat("Timepoints:\n"); print(table(tp_vec))

# ── QC ────────────────────────────────────────────────────────────────────────
seu[["percent.mt"]] <- PercentageFeatureSet(seu, pattern="^mt-")
seu <- subset(seu, subset = nFeature_RNA > 200 & nFeature_RNA < 8000 & percent.mt < 20)
cat("After QC:", ncol(seu), "cells\n")

# ── Normalise + cluster ────────────────────────────────────────────────────────
cat("Normalising...\n")
seu <- NormalizeData(seu, verbose=FALSE)
seu <- FindVariableFeatures(seu, nfeatures=2000, verbose=FALSE)
seu <- ScaleData(seu, features=VariableFeatures(seu), verbose=FALSE)
seu <- RunPCA(seu, npcs=30, verbose=FALSE)
seu <- FindNeighbors(seu, dims=1:20, verbose=FALSE)
seu <- FindClusters(seu, resolution=0.5, verbose=FALSE)
cat("Clusters:", length(unique(seu$seurat_clusters)), "\n")

# ── Cell type annotation ───────────────────────────────────────────────────────
markers_list <- list(
  Neuron          = c("Snap25","Rbfox3","Syt1","Tubb3"),
  Astrocyte       = c("Gfap","Aldh1l1","Aqp4","Slc1a3"),
  Microglia       = c("Cx3cr1","P2ry12","Tmem119","Csf1r"),
  Oligodendrocyte = c("Mbp","Plp1","Mog","Cnp"),
  OPC             = c("Pdgfra","Cspg4","Sox10","Olig2"),
  Endothelial     = c("Cldn5","Pecam1","Cdh5","Tie1"),
  Pericyte        = c("Pdgfrb","Rgs5","Notch3","Kcnj8"),
  Fibroblast      = c("Col1a1","Col1a2","Dcn","Lum")
)
avail_markers <- intersect(unique(unlist(markers_list)), rownames(seu))
Idents(seu) <- "seurat_clusters"
avg_cl <- AverageExpression(seu, features=avail_markers, assays="RNA")$RNA
colnames(avg_cl) <- sub("^g","", colnames(avg_cl))   # strip Seurat v5 "g" prefix

cluster_scores <- sapply(names(markers_list), function(ct) {
  m <- intersect(markers_list[[ct]], rownames(avg_cl))
  if(length(m)>0) colMeans(avg_cl[m,,drop=FALSE]) else rep(0, ncol(avg_cl))
})
cluster_celltype <- apply(cluster_scores, 1, function(x) names(which.max(x)))
seu@meta.data$cell_type <- cluster_celltype[as.character(seu@meta.data$seurat_clusters)]
cat("\nCell type distribution:\n"); print(sort(table(seu@meta.data$cell_type), decreasing=TRUE))
cat("\nCell type x timepoint:\n"); print(table(seu@meta.data$cell_type, seu@meta.data$timepoint))
```

Extract GSE227651 TAR and build per-timepoint Seurat object

```python

suppressPackageStartupMessages({
  library(Seurat)
  library(Matrix)
  library(dplyr)
})

# Extract TAR
cat("Extracting...\n")
system("tar -xf /workspace/geo_downloads/GSE227651/GSE227651_RAW.tar -C /workspace/geo_downloads/GSE227651/")
cat("Extracted files:\n")
print(list.files("/workspace/geo_downloads/GSE227651/", pattern="\\.gz$"))

# Load each timepoint as a separate count matrix
timepoints <- list(
  D1   = list(prefix="GSM7104632_1d"),
  D3   = list(prefix="GSM7104633_3d"),
  D7   = list(prefix="GSM7104634_7d"),
  Sham = list(prefix="GSM7104635_sham")
)

base_dir <- "/workspace/geo_downloads/GSE227651/"

seurat_list <- list()
for(tp in names(timepoints)) {
  pfx <- timepoints[[tp]]$prefix
  cat(sprintf("\nLoading %s (%s)...\n", tp, pfx))
  
  mat_tp <- readMM(gzfile(file.path(base_dir, paste0(pfx, "-matrix.mtx.gz"))))
  genes_tp <- read.table(gzfile(file.path(base_dir, paste0(pfx, "-features.tsv.gz"))),
                         header=FALSE, sep="\t")
  cells_tp <- read.table(gzfile(file.path(base_dir, paste0(pfx, "-barcodes.tsv.gz"))),
                         header=FALSE)$V1
  
  # Features file: col1=Ensembl, col2=gene symbol
  gene_names <- genes_tp$V2
  rownames(mat_tp) <- make.unique(gene_names)
  colnames(mat_tp) <- paste0(cells_tp, "_", tp)  # make barcodes unique across timepoints
  
  cat(sprintf("  %s: %d genes x %d cells\n", tp, nrow(mat_tp), ncol(mat_tp)))
  
  seu_tp <- CreateSeuratObject(counts=mat_tp, min.cells=3, min.features=200,
                               project=paste0("GSE227651_", tp))
  seu_tp@meta.data$timepoint  <- tp
  seu_tp@meta.data$condition  <- ifelse(tp=="Sham","Sham","Stroke")
  seurat_list[[tp]] <- seu_tp
}

# Merge all timepoints
cat("\nMerging...\n")
seu227 <- merge(seurat_list[[1]],
                y=seurat_list[-1],
                add.cell.ids=NULL,
                project="GSE227651")
cat("Merged:", ncol(seu227), "cells,", nrow(seu227), "genes\n")
cat("Timepoints:\n"); print(table(seu227@meta.data$timepoint))
```

```python

suppressPackageStartupMessages({library(dplyr); library(tidyr); library(jsonlite)})

# ── 1. Heatmap data ───────────────────────────────────────────────────────────
scores <- read.csv("/mnt/results/06_cell_death/core_gene_scores_all_datasets.csv")

# Dataset time-bin ordering for heatmap x-axis
time_order <- c("Acute","Multi","Chronic")
scores$time_bin <- factor(scores$time_bin, levels=time_order)
scores <- scores %>% arrange(time_bin, dataset)

heatmap_data <- scores %>%
  select(dataset, cell_type, pathway, delta_z, delta, weight, time_bin, journal_tier) %>%
  rename(z=delta_z, raw_delta=delta)

write_json(heatmap_data, "/workspace/stroke-explorer/data/heatmap.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("heatmap.json:", nrow(heatmap_data), "rows\n")

# ── 2. Forest plot data (weighted mean ± 95% CI) ─────────────────────────────
forest_data <- scores %>%
  group_by(cell_type, pathway) %>%
  summarise(
    n_datasets = n(),
    weighted_mean_z = sum(delta_z * weight) / sum(weight),
    # Weighted SE
    weighted_var = sum(weight * (delta_z - weighted_mean_z)^2) / 
                   (sum(weight) - sum(weight^2)/sum(weight)),
    weighted_se = sqrt(weighted_var / n_datasets),
    ci95_lo = weighted_mean_z - 1.96 * weighted_se,
    ci95_hi = weighted_mean_z + 1.96 * weighted_se,
    weighted_mean_raw = sum(delta * weight) / sum(weight),
    .groups="drop"
  ) %>%
  mutate(across(where(is.numeric), ~round(.x, 4)))

write_json(forest_data, "/workspace/stroke-explorer/data/forest.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("forest.json:", nrow(forest_data), "rows\n")
print(forest_data %>% filter(cell_type %in% c("Endothelial","Pericyte","Astrocyte")))

# ── 3. Temporal dynamics data ─────────────────────────────────────────────────
# Use GSE197341 (D3/D7/D14/D28/Sham) and GSE279666 (D3/D14/Sham)
ae197 <- read.csv("/mnt/results/06_cell_death/GSE197341_avg_expr_per_timepoint.csv")
ae279 <- read.csv("/mnt/results/06_cell_death/GSE279666_avg_expr_pathway_genes.csv")

sigs <- read.csv("/mnt/results/06_cell_death/cell_death_gene_signatures.csv")
core_genes <- sigs %>% filter(core==TRUE)

# Helper: compute mean pathway score per cell_type × timepoint from avg expr
compute_temporal <- function(ae, dataset_id, timepoint_map) {
  # timepoint_map: named list of col_suffix -> display_timepoint
  data_cols <- colnames(ae)[!colnames(ae) %in% c("gene","pathway")]
  
  results <- list()
  for(ct in unique(sub("\\.(D[0-9]+|Sham|Control|Stroke.*|MCAO.*)", "", data_cols))) {
    for(pathway in c("Apoptosis","Ferroptosis","Necroptosis")) {
      pathway_genes <- core_genes$gene[core_genes$pathway==pathway]
      ae_sub <- ae[ae$gene %in% pathway_genes, ]
      
      for(tp_col in names(timepoint_map)) {
        col_name <- paste0(ct, ".", tp_col)
        if(col_name %in% colnames(ae)) {
          vals <- ae_sub[[col_name]]
          results[[length(results)+1]] <- data.frame(
            dataset=dataset_id, cell_type=ct, pathway=pathway,
            timepoint_col=tp_col, timepoint=timepoint_map[[tp_col]],
            days=as.numeric(gsub("D","",tp_col)),
            mean_expr=mean(vals, na.rm=TRUE),
            stringsAsFactors=FALSE
          )
        }
      }
    }
  }
  bind_rows(results)
}

# GSE197341 timepoints
tp_map197 <- list(Sham=0, D3=3, D7=7, D14=14, D28=28)
# Need numeric days: Sham=0
tp_map197_named <- c(Sham="Sham", D3="D3", D7="D7", D14="D14", D28="D28")

# Manual approach for GSE197341
ae197_long <- list()
data_cols197 <- colnames(ae197)[!colnames(ae197) %in% c("gene","pathway")]
cell_types197 <- unique(sub("\\.(D[0-9]+|Sham)$", "", data_cols197))
tp_days <- c(Sham=0, D3=3, D7=7, D14=14, D28=28)

for(ct in cell_types197) {
  for(pw in c("Apoptosis","Ferroptosis","Necroptosis")) {
    pw_genes <- core_genes$gene[core_genes$pathway==pw]
    ae_sub <- ae197[ae197$gene %in% pw_genes, ]
    for(tp in names(tp_days)) {
      col <- paste0(ct, ".", tp)
      if(col %in% colnames(ae197)) {
        ae197_long[[length(ae197_long)+1]] <- data.frame(
          dataset="GSE197341", cell_type=ct, pathway=pw,
          timepoint=tp, days=tp_days[tp],
          mean_expr=mean(ae_sub[[col]], na.rm=TRUE),
          stringsAsFactors=FALSE
        )
      }
    }
  }
}
ae197_long <- bind_rows(ae197_long)

# GSE279666: columns like CellType..D3, CellType..D14, CellType..Sham, CellType..Control
data_cols279 <- colnames(ae279)[!colnames(ae279) %in% c("gene","pathway")]
cat("\nGSE279666 cols sample:", paste(head(data_cols279, 8), collapse=", "), "\n")

ae279_long <- list()
cell_types279 <- unique(sub("\\.\\..*$", "", data_cols279))
tp_days279 <- c(Sham=0, Control=0, D3=3, D14=14)

for(ct in cell_types279) {
  for(pw in c("Apoptosis","Ferroptosis","Necroptosis")) {
    pw_genes <- core_genes$gene[core_genes$pathway==pw]
    ae_sub <- ae279[ae279$gene %in% pw_genes, ]
    for(tp in names(tp_days279)) {
      col <- paste0(ct, "..", tp)
      if(col %in% colnames(ae279)) {
        ae279_long[[length(ae279_long)+1]] <- data.frame(
          dataset="GSE279666", cell_type=ct, pathway=pw,
          timepoint=tp, days=tp_days279[tp],
          mean_expr=mean(ae_sub[[col]], na.rm=TRUE),
          stringsAsFactors=FALSE
        )
      }
    }
  }
}
ae279_long <- bind_rows(ae279_long)

temporal_data <- bind_rows(ae197_long, ae279_long) %>%
  mutate(mean_expr=round(mean_expr, 5))

write_json(temporal_data, "/workspace/stroke-explorer/data/temporal.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("temporal.json:", nrow(temporal_data), "rows\n")
cat("Datasets:", unique(temporal_data$dataset), "\n")
cat("Cell types:", unique(temporal_data$cell_type), "\n")
```

Build per-gene expression JSON with correct separators for all datasets

```python

suppressPackageStartupMessages({library(dplyr); library(jsonlite)})

# Robust parser: split on first occurrence of separator pattern
parse_avg_expr_v2 <- function(filepath, dataset_id, sep_regex, ctrl_pattern) {
  ae <- read.csv(filepath)
  # Drop spurious "X" column if present
  ae <- ae[, !colnames(ae) %in% "X"]
  data_cols <- colnames(ae)[!colnames(ae) %in% c("gene","pathway")]
  
  result_rows <- vector("list", length(data_cols) * nrow(ae))
  idx <- 1L
  
  for(col in data_cols) {
    # Split on separator, take first part as cell_type, rest as condition
    m <- regexpr(sep_regex, col)
    if(m[1] == -1) next  # skip if no separator found
    ct <- substr(col, 1, m[1]-1)
    cond <- substr(col, m[1]+attr(m,"match.length"), nchar(col))
    is_ctrl <- grepl(ctrl_pattern, cond, ignore.case=TRUE)
    
    vals <- ae[[col]]
    genes <- ae$gene
    pathways <- ae$pathway
    
    for(i in seq_along(vals)) {
      result_rows[[idx]] <- list(
        dataset=dataset_id, gene=genes[i], pathway=pathways[i],
        cell_type=ct, condition=cond, is_control=is_ctrl,
        mean_expr=round(vals[i], 5)
      )
      idx <- idx + 1L
    }
  }
  result_rows <- result_rows[seq_len(idx-1)]
  
  # Convert to data frame efficiently
  data.frame(
    dataset  = vapply(result_rows, `[[`, character(1), "dataset"),
    gene     = vapply(result_rows, `[[`, character(1), "gene"),
    pathway  = vapply(result_rows, `[[`, character(1), "pathway"),
    cell_type= vapply(result_rows, `[[`, character(1), "cell_type"),
    condition= vapply(result_rows, `[[`, character(1), "condition"),
    is_control=vapply(result_rows, `[[`, logical(1),   "is_control"),
    mean_expr= vapply(result_rows, `[[`, numeric(1),   "mean_expr"),
    stringsAsFactors=FALSE
  )
}

# Config: filepath, dataset_id, sep_regex, ctrl_pattern
file_configs <- list(
  list("/mnt/results/06_cell_death/GSE174574_avg_expr_pathway_genes.csv",
       "GSE174574", "_", "Sham"),
  list("/mnt/results/06_cell_death/GSE197341_avg_expr_per_timepoint.csv",
       "GSE197341", "\\.", "Sham"),
  list("/mnt/results/06_cell_death/GSE227651_avg_expr_per_timepoint.csv",
       "GSE227651", "\\.", "Sham"),
  list("/mnt/results/06_cell_death/GSE244576_avg_expr_pathway_genes.csv",
       "GSE244576", "\\.\\.", "Sham"),
  list("/mnt/results/06_cell_death/GSE250245_avg_expr_pathway_genes.csv",
       "GSE250245", "__", "Sham"),
  list("/mnt/results/06_cell_death/GSE276202_avg_expr_pathway_genes.csv",
       "GSE276202", "\\.\\.", "Control"),
  list("/mnt/results/06_cell_death/GSE279666_avg_expr_pathway_genes.csv",
       "GSE279666", "\\.\\.", "Sham|Control"),
  list("/mnt/results/06_cell_death/GSE300564_avg_expr_pathway_genes.csv",
       "GSE300564", "\\.\\.", "Sham"),
  list("/mnt/results/06_cell_death/GSE310324_avg_expr_pathway_genes.csv",
       "GSE310324", "\\.\\.", "Sham")
)

all_gene_data <- list()
for(cfg in file_configs) {
  cat("Processing", cfg[[2]], "...\n")
  df <- parse_avg_expr_v2(cfg[[1]], cfg[[2]], cfg[[3]], cfg[[4]])
  cat("  Rows:", nrow(df), "| Genes:", length(unique(df$gene)),
      "| Cell types:", length(unique(df$cell_type)),
      "| Conditions:", paste(unique(df$condition), collapse=", "), "\n")
  all_gene_data[[length(all_gene_data)+1]] <- df
}

gene_data <- bind_rows(all_gene_data)
cat("\nTotal rows:", nrow(gene_data), "\n")
cat("Unique genes:", length(unique(gene_data$gene)), "\n")
cat("File size estimate:", round(object.size(gene_data)/1e6, 1), "MB\n")

write_json(gene_data, "/workspace/stroke-explorer/data/gene_expr.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("gene_expr.json written\n")
cat("File size:", round(file.info("/workspace/stroke-explorer/data/gene_expr.json")$size/1e6, 1), "MB\n")
```

```python

suppressPackageStartupMessages({library(dplyr); library(jsonlite)})

# Fix GSE250245 and GSE300564: reconstruct gene/pathway from row order
# These files have genes as row names (X column = gene names)
sigs <- read.csv("/mnt/results/06_cell_death/cell_death_gene_signatures.csv")
sig_lookup <- sigs %>% select(gene, pathway) %>% distinct()

fix_missing_gene_col <- function(filepath) {
  ae <- read.csv(filepath)
  if("gene" %in% colnames(ae)) return(ae)
  
  # X column should be gene names
  if("X" %in% colnames(ae)) {
    ae$gene <- ae$X
    ae <- ae[, !colnames(ae) %in% "X"]
    # Add pathway
    ae <- left_join(ae, sig_lookup, by="gene")
    # Move gene and pathway to front
    other_cols <- colnames(ae)[!colnames(ae) %in% c("gene","pathway")]
    ae <- ae[, c("gene","pathway", other_cols)]
    cat("  Fixed: added gene/pathway columns from X column\n")
    return(ae)
  }
  stop("Cannot fix: no X or gene column")
}

# Test fix
ae250_fixed <- fix_missing_gene_col("/mnt/results/06_cell_death/GSE250245_avg_expr_pathway_genes.csv")
cat("GSE250245 fixed: gene col present:", "gene" %in% colnames(ae250_fixed), "\n")
cat("  Sample genes:", paste(head(ae250_fixed$gene, 5), collapse=", "), "\n")
cat("  Pathway NAs:", sum(is.na(ae250_fixed$pathway)), "\n")

ae300_fixed <- fix_missing_gene_col("/mnt/results/06_cell_death/GSE300564_avg_expr_pathway_genes.csv")
cat("GSE300564 fixed: gene col present:", "gene" %in% colnames(ae300_fixed), "\n")
cat("  Pathway NAs:", sum(is.na(ae300_fixed$pathway)), "\n")

# ── Robust parser v3 ──────────────────────────────────────────────────────────
parse_avg_expr_v3 <- function(ae, dataset_id, sep_regex, ctrl_pattern) {
  ae <- ae[, !colnames(ae) %in% "X"]
  data_cols <- colnames(ae)[!colnames(ae) %in% c("gene","pathway")]
  
  all_rows <- list()
  for(col in data_cols) {
    m <- regexpr(sep_regex, col)
    if(m[1] == -1) next
    ct <- substr(col, 1, m[1]-1)
    cond <- substr(col, m[1]+attr(m,"match.length"), nchar(col))
    is_ctrl <- grepl(ctrl_pattern, cond, ignore.case=TRUE)
    
    df <- data.frame(
      dataset=dataset_id, gene=ae$gene, pathway=ae$pathway,
      cell_type=ct, condition=cond, is_control=is_ctrl,
      mean_expr=round(ae[[col]], 5),
      stringsAsFactors=FALSE
    )
    all_rows[[length(all_rows)+1]] <- df
  }
  bind_rows(all_rows)
}

# Load all files
load_ae <- function(ds) {
  if(ds %in% c("GSE197341","GSE227651")) {
    f <- paste0("/mnt/results/06_cell_death/", ds, "_avg_expr_per_timepoint.csv")
  } else {
    f <- paste0("/mnt/results/06_cell_death/", ds, "_avg_expr_pathway_genes.csv")
  }
  ae <- read.csv(f)
  if(!"gene" %in% colnames(ae)) ae <- fix_missing_gene_col(f)
  ae
}

# Config: dataset_id, sep_regex, ctrl_pattern
configs <- list(
  list("GSE174574", "_",      "Sham"),
  list("GSE197341", "\\.",    "Sham"),
  list("GSE227651", "\\.",    "Sham"),
  list("GSE244576", "\\.\\.", "Sham"),
  list("GSE250245", "__",     "Sham"),
  list("GSE276202", "\\.\\.", "Control"),
  list("GSE279666", "\\.\\.", "Sham|Control"),
  list("GSE300564", "\\.\\.", "Sham"),
  list("GSE310324", "\\.\\.", "Sham")
)

all_gene_data <- list()
for(cfg in configs) {
  ds <- cfg[[1]]
  cat("Processing", ds, "...\n")
  ae <- load_ae(ds)
  df <- parse_avg_expr_v3(ae, ds, cfg[[2]], cfg[[3]])
  cat("  Rows:", nrow(df), "| Genes:", length(unique(df$gene)),
      "| Cell types:", length(unique(df$cell_type)), "\n")
  all_gene_data[[length(all_gene_data)+1]] <- df
}

gene_data <- bind_rows(all_gene_data)
# Remove rows with NA pathway (genes not in our signature list)
gene_data <- gene_data[!is.na(gene_data$pathway), ]
cat("\nTotal rows:", nrow(gene_data), "\n")
cat("Unique genes:", length(unique(gene_data$gene)), "\n")

write_json(gene_data, "/workspace/stroke-explorer/data/gene_expr.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("gene_expr.json written\n")
cat("File size:", round(file.info("/workspace/stroke-explorer/data/gene_expr.json")$size/1e6, 1), "MB\n")
```

```python

suppressPackageStartupMessages({library(dplyr); library(jsonlite)})

# ── ISSUE 1: temporal.json EC Apoptosis direction is WRONG ───────────────────
# Expected: D3 > Sham (from Casp3 check earlier: D3=0.131 vs Sham=0.074)
# Observed: D3=0.456 < Sham=0.492 — OPPOSITE direction
# Root cause: temporal.json uses MEAN of ALL core apoptosis genes (including Bcl2l11, Bid etc.)
# which are anti-apoptotic or constitutively expressed, diluting the signal
# The gene_expr.json has the correct per-gene values

cat("=== ISSUE 1: Temporal data direction check ===\n")
tp <- fromJSON("/workspace/stroke-explorer/data/temporal.json")
ge <- fromJSON("/workspace/stroke-explorer/data/gene_expr.json")

# What genes are in the temporal mean?
core_apo <- c("Casp3","Casp7","Casp8","Casp9","Bax","Bcl2l11","Bid","Cycs","Fas","Apaf1")
ae197 <- read.csv("/mnt/results/06_cell_death/GSE197341_avg_expr_per_timepoint.csv")
ec_apo_genes <- ae197[ae197$gene %in% core_apo, grep("Endothelial", colnames(ae197), value=TRUE)]
cat("Per-gene EC apoptosis values (GSE197341):\n")
rownames(ec_apo_genes) <- ae197$gene[ae197$gene %in% core_apo]
print(round(ec_apo_genes, 4))
cat("\nMean across all 10 genes:\n")
print(round(colMeans(ec_apo_genes, na.rm=TRUE), 4))
cat("\nNote: Bcl2l11 (BIM) and Cycs are high-expression housekeeping-like genes\n")
cat("that dominate the mean and mask the Casp3/Casp9 signal\n")

# ── ISSUE 2: GSE310324 only has 2 cell types ─────────────────────────────────
cat("\n=== ISSUE 2: GSE310324 cell type coverage ===\n")
ae310 <- read.csv("/mnt/results/06_cell_death/GSE310324_avg_expr_pathway_genes.csv")
cat("GSE310324 columns:", paste(colnames(ae310), collapse=", "), "\n")
# Only Macrophage and Microglia — this is a real data limitation

# ── ISSUE 3: T.cell and B.cell in forest plot with NA CIs ────────────────────
cat("\n=== ISSUE 3: T.cell/B.cell in forest plot ===\n")
fo <- fromJSON("/workspace/stroke-explorer/data/forest.json")
rare_cts <- fo[fo$cell_type %in% c("T.cell","B.cell","NK_cell"), ]
print(rare_cts[, c("cell_type","pathway","n_datasets","weighted_mean_z","ci95_lo","ci95_hi")])
cat("\nThese appear in only 1 dataset (GSE300564) - CIs are NA because n=1\n")
cat("Recommendation: flag these as 'single-dataset' in the app\n")

# ── ISSUE 4: Temporal mean expr scale vs gene_expr scale ─────────────────────
cat("\n=== ISSUE 4: Scale consistency check ===\n")
# temporal.json uses mean of core genes (normalized counts scale ~0.4-0.6)
# gene_expr.json uses individual gene values (same scale)
# These are consistent - both from AverageExpression (log-normalized)
cat("Temporal EC Apoptosis D3:", 
    tp$mean_expr[tp$dataset=="GSE197341" & tp$cell_type=="Endothelial" & 
                 tp$pathway=="Apoptosis" & tp$timepoint=="D3"], "\n")
cat("gene_expr Casp3 EC D3:", 
    ge$mean_expr[ge$dataset=="GSE197341" & ge$cell_type=="Endothelial" & 
                 ge$condition=="D3" & ge$gene=="Casp3"], "\n")
cat("Both are log-normalized counts from AverageExpression - consistent scale\n")
```

Fix temporal.json: use effector gene subset instead of full pathway mean

```python

suppressPackageStartupMessages({library(dplyr); library(jsonlite)})

# Use biologically meaningful effector subsets for temporal summary
# These are the genes that actually execute cell death (not regulators/housekeeping)
effector_genes <- list(
  Apoptosis   = c("Casp3","Casp7","Casp9","Bax","Bid"),      # executioner caspases + pro-apoptotic
  Ferroptosis = c("Gpx4","Acsl4","Ptgs2","Tfrc","Slc7a11"),  # GPX4 axis + lipid peroxidation
  Necroptosis = c("Mlkl","Ripk3","Zbp1","Ripk1")             # MLKL/RIPK3 necrosome
)

cat("Effector gene subsets:\n")
for(pw in names(effector_genes)) cat(" ", pw, ":", paste(effector_genes[[pw]], collapse=", "), "\n")

# Rebuild temporal.json with effector genes
ae197 <- read.csv("/mnt/results/06_cell_death/GSE197341_avg_expr_per_timepoint.csv")
ae279 <- read.csv("/mnt/results/06_cell_death/GSE279666_avg_expr_pathway_genes.csv")

data_cols197 <- colnames(ae197)[!colnames(ae197) %in% c("gene","pathway")]
cell_types197 <- unique(sub("\\.(D[0-9]+|Sham)$", "", data_cols197))
tp_days197 <- c(Sham=0, D3=3, D7=7, D14=14, D28=28)

ae197_long <- list()
for(ct in cell_types197) {
  for(pw in names(effector_genes)) {
    pw_genes <- effector_genes[[pw]]
    ae_sub <- ae197[ae197$gene %in% pw_genes, ]
    for(tp in names(tp_days197)) {
      col <- paste0(ct, ".", tp)
      if(col %in% colnames(ae197)) {
        ae197_long[[length(ae197_long)+1]] <- data.frame(
          dataset="GSE197341", cell_type=ct, pathway=pw,
          timepoint=tp, days=tp_days197[tp],
          mean_expr=round(mean(ae_sub[[col]], na.rm=TRUE), 5),
          n_genes=sum(pw_genes %in% ae_sub$gene),
          stringsAsFactors=FALSE
        )
      }
    }
  }
}
ae197_long <- bind_rows(ae197_long)

data_cols279 <- colnames(ae279)[!colnames(ae279) %in% c("gene","pathway")]
cell_types279 <- unique(sub("\\.\\.(D[0-9]+|Sham|Control|Stroke.*)$", "", data_cols279))
tp_days279 <- c(Sham=0, Control=0, D3=3, D14=14)

ae279_long <- list()
for(ct in cell_types279) {
  for(pw in names(effector_genes)) {
    pw_genes <- effector_genes[[pw]]
    ae_sub <- ae279[ae279$gene %in% pw_genes, ]
    for(tp in names(tp_days279)) {
      col <- paste0(ct, "..", tp)
      if(col %in% colnames(ae279)) {
        ae279_long[[length(ae279_long)+1]] <- data.frame(
          dataset="GSE279666", cell_type=ct, pathway=pw,
          timepoint=tp, days=tp_days279[tp],
          mean_expr=round(mean(ae_sub[[col]], na.rm=TRUE), 5),
          n_genes=sum(pw_genes %in% ae_sub$gene),
          stringsAsFactors=FALSE
        )
      }
    }
  }
}
ae279_long <- bind_rows(ae279_long)

temporal_data <- bind_rows(ae197_long, ae279_long)

# Verify direction is now correct
ec_apo_197 <- temporal_data[temporal_data$dataset=="GSE197341" & 
                             temporal_data$cell_type=="Endothelial" & 
                             temporal_data$pathway=="Apoptosis",
                             c("timepoint","days","mean_expr","n_genes")]
cat("\nEC Apoptosis GSE197341 (effector genes: Casp3/7/9/Bax/Bid):\n")
print(ec_apo_197[order(ec_apo_197$days), ])
cat("D3 > Sham:", ec_apo_197$mean_expr[ec_apo_197$timepoint=="D3"] > 
                  ec_apo_197$mean_expr[ec_apo_197$timepoint=="Sham"], "\n")

# Also check Pericyte necroptosis (should be elevated)
pc_necro_197 <- temporal_data[temporal_data$dataset=="GSE197341" & 
                               temporal_data$cell_type=="Pericyte" & 
                               temporal_data$pathway=="Necroptosis",
                               c("timepoint","days","mean_expr")]
cat("\nPericyte Necroptosis GSE197341:\n")
print(pc_necro_197[order(pc_necro_197$days), ])

write_json(temporal_data, "/workspace/stroke-explorer/data/temporal.json",
           auto_unbox=TRUE, pretty=FALSE)
cat("\ntemporal.json rebuilt:", nrow(temporal_data), "rows\n")
```

```python

# Write the complete index.html file in parts
html_part1 = '''<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Stroke Cell Death Explorer</title>
  <script src="https://cdn.plot.ly/plotly-2.27.0.min.js"></script>
  <style>
    /* ===== CSS RESET & BASE ===== */
    *, *::before, *::after { box-sizing: border-box; margin: 0; padding: 0; }
    :root {
      --blue:    #0279EE;
      --green:   #75A025;
      --orange:  #FF9400;
      --yellow:  #E9ED4C;
      --pink:    #FD9BED;
      --black:   #000000;
      --bg:      #ECE9E2;
      --white:   #FFFFFF;
      --sidebar-w: 220px;
      --nav-h:   48px;
      --text:    #1a1a1a;
      --muted:   #666;
      --border:  #d0cdc6;
      --card-bg: #FFFFFF;
      --acute-c:   #0279EE;
      --multi-c:   #9B59B6;
      --chronic-c: #FF9400;
    }
    html, body { height: 100%; font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, sans-serif; background: var(--bg); color: var(--text); font-size: 14px; }

    /* ===== LOADING OVERLAY ===== */
    #loading-overlay {
      position: fixed; inset: 0; background: rgba(236,233,226,0.97);
      display: flex; flex-direction: column; align-items: center; justify-content: center;
      z-index: 9999; gap: 16px;
    }
    #loading-overlay h2 { font-size: 1.4rem; color: var(--blue); }
    #loading-overlay p  { color: var(--muted); font-size: 0.9rem; }
    .spinner {
      width: 48px; height: 48px;
      border: 5px solid var(--border);
      border-top-color: var(--blue);
      border-radius: 50%;
      animation: spin 0.8s linear infinite;
    }
    @keyframes spin { to { transform: rotate(360deg); } }
    #load-progress { width: 260px; height: 6px; background: var(--border); border-radius: 3px; overflow: hidden; }
    #load-bar { height: 100%; width: 0%; background: var(--blue); transition: width 0.3s; border-radius: 3px; }
    #load-status { font-size: 0.8rem; color: var(--muted); }

    /* ===== TOP NAV ===== */
    #topnav {
      position: fixed; top: 0; left: 0; right: 0; height: var(--nav-h);
      background: var(--white); border-bottom: 1px solid var(--border);
      display: flex; align-items: center; z-index: 100; padding: 0 12px;
      box-shadow: 0 1px 4px rgba(0,0,0,0.08);
    }
    #topnav .brand { font-weight: 700; font-size: 1rem; color: var(--blue); margin-right: 24px; white-space: nowrap; }
    #tab-bar { display: flex; gap: 2px; overflow-x: auto; }
    .tab-btn {
      padding: 8px 14px; border: none; background: transparent; cursor: pointer;
      font-size: 0.85rem; font-weight: 500; color: var(--muted); border-radius: 6px;
      white-space: nowrap; transition: background 0.15s, color 0.15s;
    }
    .tab-btn:hover { background: var(--bg); color: var(--text); }
    .tab-btn.active { background: var(--blue); color: #fff; }

    /* ===== MAIN LAYOUT ===== */
    #app {
      display: flex; margin-top: var(--nav-h); height: calc(100vh - var(--nav-h));
    }

    /* ===== SIDEBAR ===== */
    #sidebar {
      width: var(--sidebar-w); min-width: var(--sidebar-w);
      background: var(--white); border-right: 1px solid var(--border);
      display: flex; flex-direction: column; overflow-y: auto;
      padding: 16px 12px; gap: 16px;
    }
    .sidebar-title { font-size: 1rem; font-weight: 700; color: var(--blue); line-height: 1.2; }
    .sidebar-subtitle { font-size: 0.75rem; color: var(--muted); margin-top: 2px; }
    .sidebar-section { display: flex; flex-direction: column; gap: 8px; }
    .sidebar-section label { font-size: 0.78rem; font-weight: 600; color: var(--muted); text-transform: uppercase; letter-spacing: 0.04em; }
    .sidebar-section select, .sidebar-section input[type=text] {
      width: 100%; padding: 6px 8px; border: 1px solid var(--border); border-radius: 6px;
      font-size: 0.82rem; background: var(--bg); color: var(--text); outline: none;
    }
    .sidebar-section select:focus, .sidebar-section input:focus { border-color: var(--blue); }
    .check-group { display: flex; flex-direction: column; gap: 5px; }
    .check-item { display: flex; align-items: center; gap: 6px; font-size: 0.82rem; cursor: pointer; }
    .check-item input[type=checkbox] { accent-color: var(--blue); width: 14px; height: 14px; }
    .radio-group { display: flex; flex-direction: column; gap: 5px; }
    .radio-item { display: flex; align-items: center; gap: 6px; font-size: 0.82rem; cursor: pointer; }
    .radio-item input[type=radio] { accent-color: var(--blue); }
    .sidebar-divider { border: none; border-top: 1px solid var(--border); }
    .sidebar-spacer { flex: 1; }

    /* Time bin legend */
    .legend-row { display: flex; align-items: center; gap: 6px; font-size: 0.78rem; margin-bottom: 3px; }
    .legend-dot { width: 10px; height: 10px; border-radius: 50%; flex-shrink: 0; }

    /* About section */
    .about-text { font-size: 0.75rem; color: var(--muted); line-height: 1.5; }

    /* ===== CONTENT AREA ===== */
    #content { flex: 1; overflow-y: auto; padding: 20px; }
    .tab-panel { display: none; }
    .tab-panel.active { display: block; }

    /* ===== OVERVIEW TAB ===== */
    .cards-grid {
      display: grid; grid-template-columns: repeat(3, 1fr); gap: 16px;
    }
    .dataset-card {
      background: var(--white); border-radius: 10px; padding: 16px;
      border: 1px solid var(--border); box-shadow: 0 1px 4px rgba(0,0,0,0.06);
      display: flex; flex-direction: column; gap: 10px;
    }
    .card-header { display: flex; align-items: flex-start; justify-content: space-between; gap: 8px; }
    .card-id { font-size: 1.05rem; font-weight: 700; color: var(--blue); }
    .card-id a { color: inherit; text-decoration: none; }
    .card-id a:hover { text-decoration: underline; }
    .card-badges { display: flex; gap: 5px; flex-wrap: wrap; }
    .badge {
      display: inline-block; padding: 2px 8px; border-radius: 12px;
      font-size: 0.72rem; font-weight: 600; white-space: nowrap;
    }
    .badge-green  { background: #d4edda; color: #155724; }
    .badge-yellow { background: #fff3cd; color: #856404; }
    .badge-orange { background: #ffe5cc; color: #7a3e00; }
    .badge-acute   { background: #cce5ff; color: #004085; }
    .badge-multi   { background: #e8d5f5; color: #4a1a7a; }
    .badge-chronic { background: #ffe5cc; color: #7a3e00; }
    .card-meta { font-size: 0.8rem; color: var(--muted); }
    .card-meta strong { color: var(--text); }
    .chips-row { display: flex; flex-wrap: wrap; gap: 4px; }
    .chip {
      padding: 2px 8px; background: var(--bg); border-radius: 10px;
      font-size: 0.72rem; color: var(--text); border: 1px solid var(--border);
    }
    .ct-dots { display: flex; flex-wrap: wrap; gap: 6px; }
    .ct-dot-item { display: flex; align-items: center; gap: 4px; font-size: 0.75rem; }
    .ct-dot { width: 8px; height: 8px; border-radius: 50%; flex-shrink: 0; }
    .card-notes { font-size: 0.78rem; color: var(--muted); font-style: italic; border-top: 1px solid var(--border); padding-top: 8px; }

    /* ===== CHART PANELS ===== */
    .chart-panel { background: var(--white); border-radius: 10px; padding: 16px; border: 1px solid var(--border); box-shadow: 0 1px 4px rgba(0,0,0,0.06); }
    .chart-title { font-size: 1rem; font-weight: 700; margin-bottom: 4px; }
    .chart-subtitle { font-size: 0.78rem; color: var(--muted); margin-bottom: 12px; }
    .chart-note { font-size: 0.75rem; color: var(--muted); margin-top: 8px; font-style: italic; }
    .warn-note {
      display: inline-flex; align-items: center; gap: 4px;
      background: #fff3cd; border: 1px solid #ffc107; border-radius: 6px;
      padding: 4px 10px; font-size: 0.75rem; color: #856404; margin-bottom: 10px;
    }

    /* Gene search */
    .gene-search-row { display: flex; gap: 8px; align-items: center; margin-bottom: 12px; }
    .gene-search-row input { flex: 1; padding: 7px 10px; border: 1px solid var(--border); border-radius: 6px; font-size: 0.85rem; }
    .gene-search-row input:focus { outline: none; border-color: var(--blue); }
    .gene-pathway-badge { padding: 3px 10px; border-radius: 12px; font-size: 0.78rem; font-weight: 600; }
    .pathway-apoptosis  { background: #fde8e8; color: #c0392b; }
    .pathway-ferroptosis { background: #d5f5e3; color: #1a7a4a; }
    .pathway-necroptosis { background: #e8d5f5; color: #6c3483; }

    /* ===== RESPONSIVE ===== */
    @media (max-width: 900px) {
      .cards-grid { grid-template-columns: repeat(2, 1fr); }
    }
    @media (max-width: 680px) {
      #sidebar { display: none; }
      .cards-grid { grid-template-columns: 1fr; }
      #topnav .brand { display: none; }
    }
  </style>
</head>
<body>

<!-- LOADING OVERLAY -->
<div id="loading-overlay">
  <div class="spinner"></div>
  <h2>Stroke Cell Death Explorer</h2>
  <div id="load-progress"><div id="load-bar"></div></div>
  <p id="load-status">Loading datasets…</p>
</div>

<!-- TOP NAV -->
<nav id="topnav">
  <div class="brand">🧠 Stroke Explorer</div>
  <div id="tab-bar">
    <button class="tab-btn active" data-tab="overview">Overview</button>
    <button class="tab-btn" data-tab="heatmap">Heatmap</button>
    <button class="tab-btn" data-tab="forest">Forest Plot</button>
    <button class="tab-btn" data-tab="temporal">Temporal</button>
    <button class="tab-btn" data-tab="gene">Gene Explorer</button>
    <button class="tab-btn" data-tab="glial">Glial States</button>
  </div>
</nav>

<!-- MAIN APP -->
<div id="app">

  <!-- SIDEBAR -->
  <aside id="sidebar">
    <div>
      <div class="sidebar-title">Stroke Cell Death Explorer</div>
      <div class="sidebar-subtitle">9 datasets · 251 genes · Mouse MCAO</div>
    </div>

    <!-- Per-tab filter sections (shown/hidden by JS) -->

    <!-- HEATMAP filters -->
    <div id="sf-heatmap" class="sidebar-section" style="display:none">
      <label>Pathway</label>
      <select id="hm-pathway">
        <option value="All">All Pathways</option>
        <option value="Apoptosis">Apoptosis</option>
        <option value="Ferroptosis">Ferroptosis</option>
        <option value="Necroptosis">Necroptosis</option>
      </select>
      <label style="margin-top:6px">Gene Set</label>
      <div class="radio-group">
        <label class="radio-item"><input type="radio" name="hm-geneset" value="core" checked> Core genes</label>
        <label class="radio-item"><input type="radio" name="hm-geneset" value="full"> Full gene set</label>
      </div>
    </div>

    <!-- FOREST filters -->
    <div id="sf-forest" class="sidebar-section" style="display:none">
      <label>Cell Type Groups</label>
      <div class="check-group">
        <label class="check-item"><input type="checkbox" id="fp-vascular" checked> <span style="color:#0279EE;font-weight:600">●</span> Vascular</label>
        <label class="check-item"><input type="checkbox" id="fp-glial" checked> <span style="color:#75A025;font-weight:600">●</span> Glial</label>
        <label class="check-item"><input type="checkbox" id="fp-neuronal" checked> <span style="color:#FF9400;font-weight:600">●</span> Neuronal</label>
        <label class="check-item"><input type="checkbox" id="fp-immune" checked> <span style="color:#888;font-weight:600">●</span> Immune</label>
      </div>
    </div>

    <!-- TEMPORAL filters -->
    <div id="sf-temporal" class="sidebar-section" style="display:none">
      <label>Cell Type</label>
      <select id="tm-celltype">
        <option value="Endothelial">Endothelial</option>
        <option value="Astrocyte">Astrocyte</option>
        <option value="Microglia">Microglia</option>
        <option value="Neuron">Neuron</option>
        <option value="OPC">OPC</option>
        <option value="Oligodendrocyte">Oligodendrocyte</option>
        <option value="Pericyte">Pericyte</option>
        <option value="Fibroblast">Fibroblast</option>
      </select>
      <label style="margin-top:6px">Pathway</label>
      <div class="radio-group">
        <label class="radio-item"><input type="radio" name="tm-pathway" value="Apoptosis" checked> Apoptosis</label>
        <label class="radio-item"><input type="radio" name="tm-pathway" value="Ferroptosis"> Ferroptosis</label>
        <label class="radio-item"><input type="radio" name="tm-pathway" value="Necroptosis"> Necroptosis</label>
      </div>
    </div>

    <!-- GENE EXPLORER filters -->
    <div id="sf-gene" class="sidebar-section" style="display:none">
      <label>Dataset</label>
      <select id="ge-dataset"></select>
      <p style="font-size:0.72rem;color:var(--muted);margin-top:4px">Use the search box in the chart panel to select a gene.</p>
    </div>

    <!-- GLIAL STATES filters -->
    <div id="sf-glial" class="sidebar-section" style="display:none">
      <label>Dataset</label>
      <select id="gl-dataset">
        <option value="GSE227651">GSE227651</option>
        <option value="GSE174574">GSE174574</option>
        <option value="GSE279666">GSE279666</option>
      </select>
      <label style="margin-top:6px">Signatures</label>
      <div class="check-group" id="gl-sig-checks"></div>
    </div>

    <div class="sidebar-spacer"></div>
    <hr class="sidebar-divider">

    <!-- Time bin legend -->
    <div class="sidebar-section">
      <label>Time Bins</label>
      <div class="legend-row"><div class="legend-dot" style="background:var(--acute-c)"></div> Acute</div>
      <div class="legend-row"><div class="legend-dot" style="background:var(--multi-c)"></div> Multi-timepoint</div>
      <div class="legend-row"><div class="legend-dot" style="background:var(--chronic-c)"></div> Chronic</div>
    </div>

    <hr class="sidebar-divider">

    <!-- About -->
    <div class="sidebar-section">
      <label>About</label>
      <p class="about-text">
        Meta-analysis of 9 mouse MCAO scRNA-seq datasets examining apoptosis, ferroptosis, and necroptosis across 12 cell types.
        Z-scores are computed per dataset/cell-type/pathway and combined using quality-weighted averaging.
        Data covers 251 effector genes from curated pathway gene sets.
      </p>
    </div>
  </aside>

  <!-- CONTENT -->
  <main id="content">

    <!-- TAB 1: OVERVIEW -->
    <div id="tab-overview" class="tab-panel active">
      <div id="cards-grid" class="cards-grid"></div>
    </div>

    <!-- TAB 2: HEATMAP -->
    <div id="tab-heatmap" class="tab-panel">
      <div class="chart-panel">
        <div class="chart-title">Cell Death Pathway Scores Across Datasets</div>
        <div class="chart-subtitle">Z-scored mean expression of effector genes (stroke vs. control) per cell type and dataset</div>
        <div class="warn-note">⚠ GSE310324: only Macrophage + Microglia available</div>
        <div id="hm-geneset-note" style="display:none" class="warn-note">ℹ Full gene set not yet available — showing core genes</div>
        <div id="chart-heatmap" style="width:100%;min-height:520px"></div>
      </div>
    </div>

    <!-- TAB 3: FOREST PLOT -->
    <div id="tab-forest" class="tab-panel">
      <div class="chart-panel">
        <div class="chart-title">Weighted Meta-Analysis Forest Plot</div>
        <div class="chart-subtitle">Weighted mean z-score ± 95% CI across datasets; point size reflects highlight status</div>
        <div class="warn-note">⚠ GSE310324: only Macrophage + Microglia available</div>
        <div id="chart-forest" style="width:100%;min-height:500px"></div>
      </div>
    </div>

    <!-- TAB 4: TEMPORAL -->
    <div id="tab-temporal" class="tab-panel">
      <div class="chart-panel">
        <div class="chart-title">Temporal Dynamics of Cell Death Gene Expression</div>
        <div class="chart-subtitle" id="tm-subtitle">Mean of effector genes: Apoptosis (Casp3/7/9/Bax/Bid), Ferroptosis (Gpx4/Acsl4/Ptgs2/Tfrc/Slc7a11), Necroptosis (Mlkl/Ripk3/Zbp1/Ripk1)</div>
        <div id="chart-temporal" style="width:100%;min-height:420px"></div>
      </div>
    </div>

    <!-- TAB 5: GENE EXPLORER -->
    <div id="tab-gene" class="tab-panel">
      <div class="chart-panel">
        <div class="chart-title">Gene Expression Explorer</div>
        <div class="gene-search-row">
          <input type="text" id="ge-gene-input" list="gene-datalist" placeholder="Search gene… (e.g. Casp3)" value="Casp3" autocomplete="off">
          <datalist id="gene-datalist"></datalist>
          <span id="ge-pathway-badge" class="gene-pathway-badge pathway-apoptosis">Apoptosis</span>
        </div>
        <div class="chart-subtitle" id="ge-subtitle">Showing expression of <strong>Casp3</strong> across cell types and conditions</div>
        <div id="chart-gene" style="width:100%;min-height:420px"></div>
        <div class="chart-note">Expression values are log-normalized counts from AverageExpression (Seurat)</div>
      </div>
    </div>

    <!-- TAB 6: GLIAL STATES -->
    <div id="tab-glial" class="tab-panel">
      <div class="chart-panel">
        <div class="chart-title">Reactive Glia State Scores</div>
        <div class="chart-subtitle">Delta expression (stroke − control) for reactive astrocyte and microglia signatures</div>
        <div id="chart-glial" style="width:100%;min-height:420px"></div>
      </div>
    </div>

  </main>
</div>

<script>
/* ============================================================
   STROKE CELL DEATH EXPLORER — main.js (inline)
   ============================================================ */

// ── Color palette ──────────────────────────────────────────
const C = {
  blue:    '#0279EE',
  green:   '#75A025',
  orange:  '#FF9400',
  yellow:  '#E9ED4C',
  pink:    '#FD9BED',
  black:   '#000000',
  bg:      '#ECE9E2',
  white:   '#FFFFFF',
  muted:   '#888888',
  acute:   '#0279EE',
  multi:   '#9B59B6',
  chronic: '#FF9400',
};

// Cell type group membership
const CT_GROUPS = {
  vascular:  ['Endothelial','Pericyte','SMC','Fibroblast'],
  glial:     ['Astrocyte','Microglia','OPC','Oligodendrocyte'],
  neuronal:  ['Neuron'],
  immune:    ['Macrophage','T.cell','B.cell'],
};
const CT_ORDER = [
  'Endothelial','Pericyte','SMC','Fibroblast',
  'Astrocyte','Microglia','OPC','Oligodendrocyte',
  'Neuron',
  'Macrophage','T.cell','B.cell',
];
const CT_HIGHLIGHT = new Set(['Endothelial','Pericyte','Neuron','Astrocyte']);

function ctGroup(ct) {
  for (const [g, arr] of Object.entries(CT_GROUPS)) if (arr.includes(ct)) return g;
  return 'other';
}
function ctColor(ct) {
  const g = ctGroup(ct);
  return g === 'vascular' ? C.blue : g === 'glial' ? C.green : g === 'neuronal' ? C.orange : C.muted;
}

// Pathway colors
const PATHWAY_COLORS = { Apoptosis: '#C0392B', Ferroptosis: '#1A7A4A', Necroptosis: '#6C3483' };
const PATHWAY_BG     = { Apoptosis: '#fde8e8', Ferroptosis: '#d5f5e3', Necroptosis: '#e8d5f5' };

// Signature colors
const SIG_COLORS = {
  A1_markers:   '#E74C3C',
  A2_markers:   '#2ECC71',
  Pan_reactive: '#F39C12',
  M1_markers:   '#E74C3C',
  M2_markers:   '#2ECC71',
  DAM_markers:  '#9B59B6',
};
const SIG_LABELS = {
  A1_markers: 'A1', A2_markers: 'A2', Pan_reactive: 'Pan',
  M1_markers: 'M1', M2_markers: 'M2', DAM_markers: 'DAM',
};

// ── Global data store ──────────────────────────────────────
const DATA = {};

// ── Loading progress ───────────────────────────────────────
let loadedCount = 0;
const TOTAL_FILES = 6;
function updateProgress(msg) {
  loadedCount++;
  document.getElementById('load-bar').style.width = (loadedCount / TOTAL_FILES * 100) + '%';
  document.getElementById('load-status').textContent = msg;
}

// ── Fetch all data files ───────────────────────────────────
async function loadAllData() {
  const files = [
    { key: 'datasets', path: 'data/datasets.json',  msg: 'Loading datasets…' },
    { key: 'heatmap',  path: 'data/heatmap.json',   msg: 'Loading heatmap data…' },
    { key: 'forest',   path: 'data/forest.json',    msg: 'Loading forest plot data…' },
    { key: 'temporal', path: 'data/temporal.json',  msg: 'Loading temporal data…' },
    { key: 'glial',    path: 'data/glial.json',     msg: 'Loading glial state data…' },
    { key: 'gene_expr',path: 'data/gene_expr.json', msg: 'Loading gene expression data (large file)…' },
  ];

  // Load first 5 in parallel, gene_expr separately (large)
  const smallFiles = files.slice(0, 5);
  const geneFile   = files[5];

  await Promise.all(smallFiles.map(async f => {
    const r = await fetch(f.path);
    DATA[f.key] = await r.json();
    updateProgress(f.msg);
  }));

  // gene_expr with streaming note
  document.getElementById('load-status').textContent = 'Loading gene expression data (7 MB)…';
  const r = await fetch(geneFile.path);
  DATA[geneFile.key] = await r.json();
  updateProgress('Ready!');
}

// ── Tab management ─────────────────────────────────────────
let activeTab = 'overview';
const tabRendered = {};

function switchTab(tab) {
  document.querySelectorAll('.tab-btn').forEach(b => b.classList.toggle('active', b.dataset.tab === tab));
  document.querySelectorAll('.tab-panel').forEach(p => p.classList.toggle('active', p.id === 'tab-' + tab));
  document.querySelectorAll('[id^="sf-"]').forEach(s => s.style.display = 'none');
  const sf = document.getElementById('sf-' + tab);
  if (sf) sf.style.display = 'flex';
  activeTab = tab;
  renderTab(tab);
}

function renderTab(tab) {
  if (tab === 'overview')  renderOverview();
  if (tab === 'heatmap')   renderHeatmap();
  if (tab === 'forest')    renderForest();
  if (tab === 'temporal')  renderTemporal();
  if (tab === 'gene')      renderGene();
  if (tab === 'glial')     renderGlial();
}

// ── Utility helpers ────────────────────────────────────────
function fmtNum(n) { return n.toLocaleString(); }
function fmtPct(w) { return (w * 100).toFixed(1) + '%'; }

function timeBinBadge(tb) {
  const cls = tb === 'Acute' ? 'badge-acute' : tb === 'Multi' ? 'badge-multi' : 'badge-chronic';
  return `<span class="badge ${cls}">${tb}</span>`;
}
function weightBadge(w) {
  const pct = w * 100;
  const cls = pct > 12 ? 'badge-green' : pct >= 10 ? 'badge-yellow' : 'badge-orange';
  return `<span class="badge ${cls}">w=${fmtPct(w)}</span>`;
}

const CT_DOT_COLORS = {
  Endothelial: '#0279EE', Pericyte: '#1A5276', SMC: '#2E86C1', Fibroblast: '#85C1E9',
  Astrocyte: '#75A025', Microglia: '#1E8449', OPC: '#A9DFBF', Oligodendrocyte: '#27AE60',
  Neuron: '#FF9400',
  Macrophage: '#888', 'T.cell': '#AAA', 'B.cell': '#CCC',
};

// ── TAB 1: OVERVIEW ────────────────────────────────────────
function renderOverview() {
  const grid = document.getElementById('cards-grid');
  if (grid.children.length > 0) return; // already rendered
  grid.innerHTML = DATA.datasets.map(d => {
    const geoUrl = `https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=${d.id}`;
    const tpChips = d.timepoints.map(t => `<span class="chip">${t}</span>`).join('');
    const ctDots  = d.cell_types.map(ct => {
      const col = CT_DOT_COLORS[ct] || '#999';
      return `<span class="ct-dot-item"><span class="ct-dot" style="background:${col}"></span>${ct}</span>`;
    }).join('');
    return `
    <div class="dataset-card">
      <div class="card-header">
        <div class="card-id"><a href="${geoUrl}" target="_blank" rel="noopener">${d.id}</a></div>
        <div class="card-badges">
          ${weightBadge(d.weight)}
          ${timeBinBadge(d.time_bin)}
        </div>
      </div>
      <div class="card-meta"><strong>${d.journal}</strong> · ${d.year} · Tier ${d.journal_tier}</div>
      <div class="card-meta"><strong>${fmtNum(d.n_cells)}</strong> cells · ${d.region}</div>
      <div class="card-meta"><strong>Timepoints:</strong></div>
      <div class="chips-row">${tpChips}</div>
      <div class="card-meta"><strong>Cell types:</strong></div>
      <div class="ct-dots">${ctDots}</div>
      <div class="card-notes">${d.notes}</div>
    </div>`;
  }).join('');
}

// ── TAB 2: HEATMAP ─────────────────────────────────────────
function getHeatmapDatasets() {
  // Order: Acute first, then Multi, then Chronic; within each bin by weight desc
  const ds = DATA.datasets.slice().sort((a, b) => {
    const binOrder = { Acute: 0, Multi: 1, Chronic: 2 };
    const bo = binOrder[a.time_bin] - binOrder[b.time_bin];
    return bo !== 0 ? bo : b.weight - a.weight;
  });
  return ds.map(d => d.id);
}

function renderHeatmap() {
  const pathway = document.getElementById('hm-pathway').value;
  const pathways = pathway === 'All' ? ['Apoptosis','Ferroptosis','Necroptosis'] : [pathway];
  const dsOrder  = getHeatmapDatasets();

  // Cell types present in data
  const ctPresent = CT_ORDER.filter(ct => DATA.heatmap.some(r => r.cell_type === ct));

  // Build one subplot per pathway
  const nCols = pathways.length;
  const traces = [];
  const annotations = [];

  pathways.forEach((pw, pi) => {
    const rows = DATA.heatmap.filter(r => r.pathway === pw);
    // Build z matrix: rows=cell_types, cols=datasets
    const zMatrix = ctPresent.map(ct =>
      dsOrder.map(ds => {
        const row = rows.find(r => r.cell_type === ct && r.dataset === ds);
        return row ? row.z : null;
      })
    );
    const textMatrix = ctPresent.map(ct =>
      dsOrder.map(ds => {
        const row = rows.find(r => r.cell_type === ct && r.dataset === ds);
        if (!row) return '';
        return `Dataset: ${ds}<br>Cell type: ${ct}<br>Pathway: ${pw}<br>Z-score: ${row.z?.toFixed(3)}<br>Raw Δ: ${row.raw_delta?.toFixed(4)}`;
      })
    );

    const xref = nCols === 1 ? 'x' : `x${pi + 1}`;
    const yref = nCols === 1 ? 'y' : `y${pi + 1}`;

    traces.push({
      type: 'heatmap',
      z: zMatrix,
      x: dsOrder,
      y: ctPresent,
      text: textMatrix,
      hovertemplate: '%{text}<extra></extra>',
      colorscale: 'RdBu',
      reversescale: true,
      zmin: -2.5, zmax: 2.5,
      showscale: pi === pathways.length - 1,
      colorbar: { title: 'Z-score', thickness: 14, len: 0.8 },
      xaxis: xref,
      yaxis: yref,
    });

    // Pathway label annotation
    annotations.push({
      text: `<b>${pw}</b>`,
      x: (pi + 0.5) / nCols,
      y: 1.04,
      xref: 'paper', yref: 'paper',
      showarrow: false,
      font: { size: 13, color: PATHWAY_COLORS[pw] },
      xanchor: 'center',
    });

    // Bold labels for highlighted cell types
    ctPresent.forEach((ct, ci) => {
      if (CT_HIGHLIGHT.has(ct)) {
        annotations.push({
          text: `<b>${ct}</b>`,
          x: pi === 0 ? -0.01 : (pi / nCols) - 0.01,
          y: ci / (ctPresent.length - 1),
          xref: 'paper', yref: yref,
          showarrow: false,
          font: { size: 10, color: '#000' },
          xanchor: 'right',
        });
      }
    });
  });

  // Build subplot axes layout
  const layout = {
    annotations,
    margin: { l: 110, r: 80, t: 60, b: 100 },
    paper_bgcolor: 'white',
    plot_bgcolor: 'white',
    height: Math.max(420, ctPresent.length * 32 + 160),
  };

  if (nCols === 1) {
    layout.xaxis = { tickangle: -45, tickfont: { size: 10 }, side: 'bottom' };
    layout.yaxis = { autorange: 'reversed', tickfont: { size: 10 } };
  } else {
    const gap = 0.06;
    const w = (1 - gap * (nCols - 1)) / nCols;
    for (let i = 0; i < nCols; i++) {
      const xi = i === 0 ? 'xaxis' : `xaxis${i + 1}`;
      const yi = i === 0 ? 'yaxis' : `yaxis${i + 1}`;
      layout[xi] = {
        domain: [i * (w + gap), i * (w + gap) + w],
        tickangle: -45, tickfont: { size: 10 }, anchor: i === 0 ? 'y' : `y${i + 1}`,
      };
      layout[yi] = {
        autorange: 'reversed', tickfont: { size: 10 },
        anchor: i === 0 ? 'x' : `x${i + 1}`,
        showticklabels: i === 0,
      };
    }
  }

  Plotly.react('chart-heatmap', traces, layout, { responsive: true, displayModeBar: false });
}

// ── TAB 3: FOREST PLOT ─────────────────────────────────────
function renderForest() {
  const showVascular  = document.getElementById('fp-vascular').checked;
  const showGlial     = document.getElementById('fp-glial').checked;
  const showNeuronal  = document.getElementById('fp-neuronal').checked;
  const showImmune    = document.getElementById('fp-immune').checked;

  const groupVisible = { vascular: showVascular, glial: showGlial, neuronal: showNeuronal, immune: showImmune };

  const pathways = ['Apoptosis','Ferroptosis','Necroptosis'];
  const traces = [];
  const annotations = [];

  const gap = 0.06;
  const w = (1 - gap * 2) / 3;

  pathways.forEach((pw, pi) => {
    const rows = DATA.forest
      .filter(r => r.pathway === pw)
      .filter(r => groupVisible[ctGroup(r.cell_type)])
      .sort((a, b) => a.weighted_mean_z - b.weighted_mean_z);

    const xref = pi === 0 ? 'x' : `x${pi + 1}`;
    const yref = pi === 0 ? 'y' : `y${pi + 1}`;

    // Separate into normal (n>1) and sparse (n=1)
    const normal = rows.filter(r => r.n_datasets > 1);
    const sparse = rows.filter(r => r.n_datasets === 1);

    // Normal points with error bars
    if (normal.length) {
      traces.push({
        type: 'scatter',
        mode: 'markers',
        x: normal.map(r => r.weighted_mean_z),
        y: normal.map(r => r.cell_type),
        error_x: {
          type: 'data',
          symmetric: false,
          array:      normal.map(r => (r.ci95_hi ?? r.weighted_mean_z) - r.weighted_mean_z),
          arrayminus: normal.map(r => r.weighted_mean_z - (r.ci95_lo ?? r.weighted_mean_z)),
          color: normal.map(r => ctColor(r.cell_type)),
          thickness: 2, width: 6,
        },
        marker: {
          color: normal.map(r => ctColor(r.cell_type)),
          size:  normal.map(r => CT_HIGHLIGHT.has(r.cell_type) ? 12 : 8),
          line:  { color: normal.map(r => ctColor(r.cell_type)), width: 2 },
        },
        customdata: normal.map(r => [r.cell_type, pw, r.weighted_mean_z, r.ci95_lo, r.ci95_hi, r.n_datasets]),
        hovertemplate: '<b>%{customdata[0]}</b><br>Pathway: %{customdata[1]}<br>Weighted mean z: %{customdata[2]:.3f}<br>95% CI: [%{customdata[3]:.3f}, %{customdata[4]:.3f}]<br>n datasets: %{customdata[5]}<extra></extra>',
        showlegend: false,
        xaxis: xref, yaxis: yref,
      });
    }

    // Sparse points (n=1) — open circles, no error bars
    if (sparse.length) {
      traces.push({
        type: 'scatter',
        mode: 'markers+text',
        x: sparse.map(r => r.weighted_mean_z),
        y: sparse.map(r => r.cell_type),
        text: sparse.map(r => ' (n=1)'),
        textposition: 'middle right',
        textfont: { size: 9, color: '#888' },
        marker: {
          color: 'white',
          size: 8,
          line: { color: sparse.map(r => ctColor(r.cell_type)), width: 2 },
          symbol: 'circle-open',
        },
        customdata: sparse.map(r => [r.cell_type, pw, r.weighted_mean_z, r.n_datasets]),
        hovertemplate: '<b>%{customdata[0]}</b><br>Pathway: %{customdata[1]}<br>Weighted mean z: %{customdata[2]:.3f}<br>n datasets: %{customdata[3]}<extra></extra>',
        showlegend: false,
        xaxis: xref, yaxis: yref,
      });
    }

    // Zero line shape
    const xDom = [pi * (w + gap), pi * (w + gap) + w];
    annotations.push({
      text: `<b>${pw}</b>`,
      x: (xDom[0] + xDom[1]) / 2,
      y: 1.04,
      xref: 'paper', yref: 'paper',
      showarrow: false,
      font: { size: 13, color: PATHWAY_COLORS[pw] },
      xanchor: 'center',
    });

    const xi = pi === 0 ? 'xaxis' : `xaxis${pi + 1}`;
    const yi = pi === 0 ? 'yaxis' : `yaxis${pi + 1}`;
    const allRows = rows;
    const xVals = allRows.map(r => r.weighted_mean_z);
    const xMin = Math.min(-0.5, ...xVals.map(v => v - 1.5));
    const xMax = Math.max(0.5,  ...xVals.map(v => v + 1.5));
  });

  const layout = {
    annotations,
    margin: { l: 110, r: 30, t: 60, b: 50 },
    paper_bgcolor: 'white',
    plot_bgcolor: '#fafafa',
    height: Math.max(400, 30 * (DATA.forest.filter(r => groupVisible[ctGroup(r.cell_type)]).length / 3 + 2) + 120),
    shapes: [],
  };

  for (let i = 0; i < 3; i++) {
    const xi = i === 0 ? 'xaxis' : `xaxis${i + 1}`;
    const yi = i === 0 ? 'yaxis' : `yaxis${i + 1}`;
    const xDom = [i * (w + gap), i * (w + gap) + w];
    layout[xi] = { domain: xDom, zeroline: true, zerolinecolor: '#333', zerolinewidth: 1.5, gridcolor: '#eee', tickfont: { size: 10 }, anchor: i === 0 ? 'y' : `y${i + 1}` };
    layout[yi] = { tickfont: { size: 10 }, anchor: i === 0 ? 'x' : `x${i + 1}`, showticklabels: i === 0 };
  }

  Plotly.react('chart-forest', traces, layout, { responsive: true, displayModeBar: false });
}

// ── TAB 4: TEMPORAL ────────────────────────────────────────
function renderTemporal() {
  const cellType = document.getElementById('tm-celltype').value;
  const pathway  = document.querySelector('input[name="tm-pathway"]:checked').value;

  const rows = DATA.temporal.filter(r => r.cell_type === cellType && r.pathway === pathway);
  const ds1  = rows.filter(r => r.dataset === 'GSE197341').sort((a, b) => a.days - b.days);
  const ds2  = rows.filter(r => r.dataset === 'GSE279666').sort((a, b) => a.days - b.days);

  const col = PATHWAY_COLORS[pathway];

  // Day label formatter
  function dayLabel(d) { return d === 0 ? 'Sham' : `D${d}`; }

  const traces = [];

  if (ds1.length) {
    traces.push({
      type: 'scatter', mode: 'lines+markers',
      name: 'GSE197341',
      x: ds1.map(r => r.days),
      y: ds1.map(r => r.mean_expr),
      line: { color: col, width: 2.5, dash: 'solid' },
      marker: { color: col, size: 7 },
      customdata: ds1.map(r => [r.dataset, r.timepoint, r.days, r.mean_expr, r.n_genes]),
      hovertemplate: 'Dataset: %{customdata[0]}<br>Timepoint: %{customdata[1]}<br>Days: %{customdata[2]}<br>Mean expr: %{customdata[3]:.4f}<br>n genes: %{customdata[4]}<extra></extra>',
    });
  }
  if (ds2.length) {
    traces.push({
      type: 'scatter', mode: 'lines+markers',
      name: 'GSE279666',
      x: ds2.map(r => r.days),
      y: ds2.map(r => r.mean_expr),
      line: { color: col, width: 2.5, dash: 'dash' },
      marker: { color: col, size: 7, symbol: 'square' },
      customdata: ds2.map(r => [r.dataset, r.timepoint, r.days, r.mean_expr, r.n_genes]),
      hovertemplate: 'Dataset: %{customdata[0]}<br>Timepoint: %{customdata[1]}<br>Days: %{customdata[2]}<br>Mean expr: %{customdata[3]:.4f}<br>n genes: %{customdata[4]}<extra></extra>',
    });
  }

  // All day values for tick labels
  const allDays = [0, 3, 7, 14, 28];
  const tickVals = allDays;
  const tickText = allDays.map(dayLabel);

  const layout = {
    xaxis: {
      title: 'Days post-stroke',
      tickvals: tickVals, ticktext: tickText,
      zeroline: false, gridcolor: '#eee',
    },
    yaxis: { title: 'Mean expression (log-norm)', gridcolor: '#eee' },
    shapes: [{
      type: 'line', x0: 0, x1: 0, y0: 0, y1: 1,
      xref: 'x', yref: 'paper',
      line: { color: '#999', width: 1.5, dash: 'dot' },
    }],
    annotations: [{
      x: 0, y: 1.02, xref: 'x', yref: 'paper',
      text: 'Sham/Control', showarrow: false,
      font: { size: 10, color: '#888' }, xanchor: 'center',
    }],
    legend: { x: 0.02, y: 0.98, bgcolor: 'rgba(255,255,255,0.8)', bordercolor: '#ddd', borderwidth: 1 },
    margin: { l: 60, r: 30, t: 30, b: 60 },
    paper_bgcolor: 'white', plot_bgcolor: '#fafafa',
    height: 420,
  };

  Plotly.react('chart-temporal', traces, layout, { responsive: true, displayModeBar: false });
}

// ── TAB 5: GENE EXPLORER ───────────────────────────────────
function initGeneExplorer() {
  // Populate gene datalist
  const genes = [...new Set(DATA.gene_expr.map(r => r.gene))].sort();
  const dl = document.getElementById('gene-datalist');
  dl.innerHTML = genes.map(g => `<option value="${g}">`).join('');

  // Populate dataset selector
  const datasets = [...new Set(DATA.gene_expr.map(r => r.dataset))].sort();
  const sel = document.getElementById('ge-dataset');
  sel.innerHTML = datasets.map(d => `<option value="${d}"${d === 'GSE197341' ? ' selected' : ''}>${d}</option>`).join('');

  // Event listeners
  document.getElementById('ge-gene-input').addEventListener('input', debounce(renderGene, 300));
  sel.addEventListener('change', renderGene);
}

// Condition ordering helper
function conditionOrder(cond) {
  const order = { 'Sham': 0, 'Control': 0, 'D1': 1, 'D3': 2, 'D7': 3, 'D14': 4, 'D28': 5 };
  return order[cond] ?? 99;
}

// Stroke condition color gradient (light blue → dark blue)
const STROKE_COLORS = ['#AED6F1','#5DADE2','#2E86C1','#1A5276','#0D2B4E'];

function renderGene() {
  const gene    = document.getElementById('ge-gene-input').value.trim();
  const dataset = document.getElementById('ge-dataset').value;
  if (!gene) return;

  const rows = DATA.gene_expr.filter(r => r.gene === gene && r.dataset === dataset);
  if (!rows.length) {
    Plotly.react('chart-gene', [], { title: `No data for ${gene} in ${dataset}` }, { responsive: true });
    return;
  }

  // Pathway for this gene
  const pathway = rows[0].pathway;
  const badge = document.getElementById('ge-pathway-badge');
  badge.textContent = pathway;
  badge.className = `gene-pathway-badge pathway-${pathway.toLowerCase()}`;
  document.getElementById('ge-subtitle').innerHTML = `Showing expression of <strong>${gene}</strong> (${pathway}) across cell types and conditions in <strong>${dataset}</strong>`;

  // Get unique conditions sorted
  const conditions = [...new Set(rows.map(r => r.condition))].sort((a, b) => conditionOrder(a) - conditionOrder(b));
  const cellTypes  = CT_ORDER.filter(ct => rows.some(r => r.cell_type === ct));

  // Assign colors
  const strokeConds = conditions.filter(c => !rows.find(r => r.condition === c)?.is_control);
  const condColor = {};
  conditions.forEach(c => {
    const isCtrl = rows.find(r => r.condition === c)?.is_control;
    if (isCtrl) { condColor[c] = '#999'; }
    else {
      const idx = strokeConds.indexOf(c);
      condColor[c] = STROKE_COLORS[Math.min(idx, STROKE_COLORS.length - 1)];
    }
  });

  const traces = conditions.map(cond => {
    const condRows = rows.filter(r => r.condition === cond);
    return {
      type: 'bar',
      name: cond,
      x: cellTypes,
      y: cellTypes.map(ct => {
        const r = condRows.find(r => r.cell_type === ct);
        return r ? r.mean_expr : null;
      }),
      marker: { color: condColor[cond] },
      customdata: cellTypes.map(ct => {
        const r = condRows.find(r => r.cell_type === ct);
        return r ? [gene, ct, cond, r.mean_expr, r.is_control] : [gene, ct, cond, null, null];
      }),
      hovertemplate: 'Gene: %{customdata[0]}<br>Cell type: %{customdata[1]}<br>Condition: %{customdata[2]}<br>Mean expr: %{y:.4f}<br>Control: %{customdata[4]}<extra></extra>',
    };
  });

  const layout = {
    barmode: 'group',
    xaxis: { title: 'Cell type', tickangle: -30, tickfont: { size: 11 } },
    yaxis: { title: 'Mean expression (log-norm)', gridcolor: '#eee' },
    legend: { orientation: 'h', y: -0.25 },
    margin: { l: 60, r: 20, t: 20, b: 100 },
    paper_bgcolor: 'white', plot_bgcolor: '#fafafa',
    height: 420,
  };

  Plotly.react('chart-gene', traces, layout, { responsive: true, displayModeBar: false });
}

// ── TAB 6: GLIAL STATES ────────────────────────────────────
function initGlialStates() {
  // Build signature checkboxes
  const sigs = [...new Set(DATA.glial.map(r => r.signature))];
  const container = document.getElementById('gl-sig-checks');
  container.innerHTML = sigs.map(s => `
    <label class="check-item">
      <input type="checkbox" class="gl-sig-cb" value="${s}" checked>
      <span style="color:${SIG_COLORS[s]};font-weight:600">●</span> ${SIG_LABELS[s] || s}
    </label>`).join('');

  document.getElementById('gl-dataset').addEventListener('change', renderGlial);
  container.addEventListener('change', renderGlial);
}

function renderGlial() {
  const dataset = document.getElementById('gl-dataset').value;
  const activeSigs = [...document.querySelectorAll('.gl-sig-cb:checked')].map(cb => cb.value);

  const rows = DATA.glial.filter(r => r.dataset === dataset && activeSigs.includes(r.signature));
  const cellTypes = [...new Set(rows.map(r => r.cell_type))].sort();

  const traces = activeSigs.map(sig => {
    const sigRows = rows.filter(r => r.signature === sig);
    return {
      type: 'bar',
      name: SIG_LABELS[sig] || sig,
      x: cellTypes,
      y: cellTypes.map(ct => {
        const r = sigRows.find(r => r.cell_type === ct);
        return r ? r.delta : null;
      }),
      marker: { color: SIG_COLORS[sig] },
      customdata: cellTypes.map(ct => {
        const r = sigRows.find(r => r.cell_type === ct);
        return r ? [dataset, ct, sig, r.ctrl_mean, r.stroke_mean, r.delta, r.n_genes] : [dataset, ct, sig, null, null, null, null];
      }),
      hovertemplate: 'Dataset: %{customdata[0]}<br>Cell type: %{customdata[1]}<br>Signature: %{customdata[2]}<br>Ctrl mean: %{customdata[3]:.4f}<br>Stroke mean: %{customdata[4]:.4f}<br>Delta: %{customdata[5]:.4f}<br>n genes: %{customdata[6]}<extra></extra>',
    };
  });

  const layout = {
    barmode: 'group',
    xaxis: { title: 'Cell type', tickfont: { size: 11 } },
    yaxis: { title: 'Delta (stroke − control)', zeroline: true, zerolinecolor: '#333', gridcolor: '#eee' },
    legend: { orientation: 'h', y: -0.2 },
    margin: { l: 60, r: 20, t: 20, b: 80 },
    paper_bgcolor: 'white', plot_bgcolor: '#fafafa',
    height: 420,
  };

  Plotly.react('chart-glial', traces, layout, { responsive: true, displayModeBar: false });
}

// ── Debounce utility ───────────────────────────────────────
function debounce(fn, ms) {
  let t; return (...args) => { clearTimeout(t); t = setTimeout(() => fn(...args), ms); };
}

// ── Wire up event listeners ────────────────────────────────
function wireEvents() {
  // Tab buttons
  document.querySelectorAll('.tab-btn').forEach(btn => {
    btn.addEventListener('click', () => switchTab(btn.dataset.tab));
  });

  // Heatmap filters
  document.getElementById('hm-pathway').addEventListener('change', renderHeatmap);
  document.querySelectorAll('input[name="hm-geneset"]').forEach(r => {
    r.addEventListener('change', () => {
      const note = document.getElementById('hm-geneset-note');
      note.style.display = r.value === 'full' ? 'inline-flex' : 'none';
      renderHeatmap();
    });
  });

  // Forest filters
  ['fp-vascular','fp-glial','fp-neuronal','fp-immune'].forEach(id => {
    document.getElementById(id).addEventListener('change', renderForest);
  });

  // Temporal filters
  document.getElementById('tm-celltype').addEventListener('change', renderTemporal);
  document.querySelectorAll('input[name="tm-pathway"]').forEach(r => {
    r.addEventListener('change', renderTemporal);
  });
}

// ── App initialization ─────────────────────────────────────
async function init() {
  try {
    await loadAllData();
    document.getElementById('loading-overlay').style.display = 'none';
    wireEvents();
    initGeneExplorer();
    initGlialStates();
    renderOverview();
    // Show heatmap sidebar filter
    document.getElementById('sf-heatmap').style.display = 'none';
  } catch (err) {
    document.getElementById('load-status').textContent = 'Error loading data: ' + err.message;
    console.error(err);
  }
}

init();
</script>
</body>
</html>'''

with open('/workspace/subagents/b5b5e146/index.html', 'w') as f:
    f.write(html_part1)

print(f"Written {len(html_part1)} characters")
```
